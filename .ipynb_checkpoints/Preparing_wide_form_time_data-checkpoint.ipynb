## Preparing Wide Form Time Data

### Imports, loading, functions

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import os
# Set random seed
SEED = 321
np.random.seed(SEED)
```

```python
## Customization options
# plt.style.use(['fivethirtyeight'])#'seaborn-talk'])#,
plt.rcParams['figure.facecolor']='white'
plt.rcParams['figure.figsize']=(12,4)
# Import tick customization tools
import matplotlib.ticker as mticks
import matplotlib.dates as mdates
sns.set_context("talk", font_scale=0.9)
```

```python
def format_xdates(ax, fig=None, xtick_fontweight='bold', title=None):
    """Formats x-axis with minor ticks every 3 months, and major ticks every year.
    Includes different grid options for visibility."""
    # Create the locator to place ticks every 3 months
    loc_3months = mdates.MonthLocator(interval=3)
    fmt_months = mdates.DateFormatter("%b")
    # For major year ticks
    loc_year = mdates.YearLocator()
    fmt_year = mdates.DateFormatter("%Y")
    
    # Change the font of the major ticks to stand out
    ax.set_xticks(ax.get_xticks())
    ax.set_xticklabels(ax.get_xticklabels(), fontweight=xtick_fontweight)
    
    # Customize major ticks
    ax.xaxis.set_major_locator(loc_year)
    ax.xaxis.set_major_formatter(fmt_year)
    # Customize minor ticks
    ax.xaxis.set_minor_locator(loc_3months)
    ax.xaxis.set_minor_formatter(fmt_months)

    # Making major-minor gridlines visually distinct
    ax.grid(which='major', axis='x', color='k')
    ax.grid(which='minor', axis='x', ls=':')
    if fig is None:
        fig = ax.get_figure()
    # Rotate dates
    fig.automft_xdate(which='major', rotation=90, ha='center')
    return fig, ax 
    
```

### Introduction of Wide-form vs Long-form data

**Corona Virus Time Series - Cases**

```python
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vRtZ3XGbdzViSOsO6Gam4LQAQTSrFMXSAz_CIcBmmQYBv9uuggbGcxUuqXtMeUPiHR9NyI6At3apf5d/pub?output=csv"
df = pd.read_csv(url)
df
```

- This data is in wide-form, where each date has it's own column. It must be converted to long-form before it can be used as a Pandas time series

**Using pd.melt**
- Pandas has a function named pd.melt which turns wide-form data into long-form, but still retains some id columns.
- The columns we want to keep associated with each timestamp will be our id_vars
    - If we want EVERY other col converted to rows, we can stop here
    - If we only want SOME cols converted to rows, we would use value_vars argument

```python
# Examine existing columns
df.columns
```

- We want to keep the first four columns, so we will put those into the id_vars argument

```python
melted = pd.melt(df, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'])
melted
```

- Using melt has compressed all the dates into a column titled 'variable', and the number of cases on each date have moved to a column titled 'value'
- We can also name these columns upon creation - see below.
- (We also dropped 'Lat' and 'Long' to simplify the example)

```python
melted = pd.melt(df.drop(columns=['Lat', 'Long']),
                 id_vars = ['Province/State', 'Country/Region'],
                 value_name = 'Confirmed', var_name='Date')
melted
```

- Now that we have our melted dataframe we can begin slicing out the needed countries and preparing our datetime index.

```python
# First convert the 'Date' column values to datetime objects
melted['Date'] = pd.to_datetime(melted['Date'])
melted.info()
```

- This dataset has the same date repreated many times, one for each country. We will need to set Date as our index, but we are not resampling quite yet.

```python
melted = melted.set_index('Date')
melted
```

- What would be the expected result of resampling the data at this stage?

```python
melted.resample("D").sum()
```

- We now have the sum for each day, but have lost the division by country. 

### Combining .groupby and .resample for groupwise time series processing

#### Using groupby and resample together

```python
# How many countries does our data cover?
melted['Country/Region'].nunique()
```

- We can combine the use of .groupby and .resample("D") to get results group-wise
- By doing this, we will end with an index with two columns: a MultiIndex
    - When using .resample we must use an aggregate method, however, in this instance any agg method will return the same total so it doesn't matter which we pick. In the example here, we use .sum()

```python
df_ts = melted.groupby('Country/Region').resample("D").sum()
df_ts
```

```python
df_ts.head().index
```

- However, we have now retained our ability to locate information by country.

```python
df_ts.loc['Afghanistan']
```

#### Filtering by Date with a MultiIndex using pd.IndexSlice

- While we can easily filter by country, how we filter by date will now be different. The normal method will return an error

```python
# This will return an error
# df_ts.loc['2020']
```

```python
# Instead, use pd.IndexSlice along with .loc to filter for values in 2020
df_ts.loc[pd.IndexSlice[:, '2020'],:]
```

```python
# We can also filter by both date and country
df_ts.loc[pd.IndexSlice['Afghanistan':'Algeria', '2020'],:]
```

### Plotting with a MultiIndex

- Try plotting time series data for several countries at once - US, Mexico, and Canada

```python
countries = ['Canada', 'Mexico', 'US']
plot_df = df_ts.loc[countries].copy()
plot_df
```

```python
plot_df.plot(figsize=(12,4));
```

- Pandas is unsure what to consider a group, so it has combined all the data into one line.
- To visualize each line separately we have to modify our multi-index temporarily, which we can perform with .unstack()

#### Using unstack() for plotting with a multi-index

**Index levels**
- Currently the multi-index has 2 levels
    - Level 0: (outer-most, on the left) Country
    - Level 1: Date
- To demonstrate, use df.index.get_level_values() functions to get values from 1 index at a time

```python
# Level 0 index - Countries
plot_df.index.get_level_values(0)
```

```python
# Level 1 index - Date
plot_df.index.get_level_values(1)
```

- Documentation for Pandas .unstack() says it will "pivot a level of the index labels"
- .unstack() will allow pivoting of data in a way that keeps Date as the index, and convert the outer index (Country/Region) into columns

#### Applying unstack in the visualization

```python
# unstack accepts a 'level' argument for which the level of the multi-index should be converted into columns
# Because the goal is to convert the outermost index (level 0) to columns, specify level=0
unstacked = plot_df.unstack(level=0)
unstacked
```

- With the dataframe unstacked, there is now a separate column for each country

```python
ax = unstacked.plot()
```

- Lastly, address the legend. Each line has 'Confirmed, "Country Name"'

```python
unstacked.columns
```

- By slicing out the 'Confirmed' column from the multi-index, it's possible to bypass the outer level of the multi-index so 'Confirmed' column will not appear in the legend.

```python
ax = unstacked['Confirmed'].plot()
```
