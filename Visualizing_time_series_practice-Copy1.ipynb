## Answers

### Prep work

**1. Convert date column to datetime type**

```python
df['datetime'] = pd.to_datetime(df['date'], format='%Y%m%d')
```

**2. Set datetime as new index and inspect**

```python
df = df.set_index(['datetime'])
df.head()
```

**3. Filter down to the requested data**
- Only data after the year 2000
- preceipitation
- mean_temp
- min_temp
- max_temp
- snow_depth

```python
df = df.drop(columns=['date', 'cloud_cover', 'sunshine', 'global_radiation', 'pressure'])
df = df.loc['2000':]
df.info()
df.head()
```

**4. Impute missing values**

```python
df['max_temp'] = df['max_temp'].interpolate()
df['min_temp'] = df['min_temp'].interpolate()
df['mean_temp'] = df['mean_temp'].interpolate()
```

```python
df['precipitation'] = df['precipitation'].fillna(0)
df['snow_depth'] = df['snow_depth'].fillna(0)
```

**6. Plot precipitation and snow_depth for 2000-2020 on a single figure**

```python
# Get temporary dataframe for relevant years
df_rain = df.loc['2000':'2020']
# Get columns of relevant data for plotting
precip = df_rain['precipitation']
snow_depth = df_rain['snow_depth']
```

```python
# Set figure parameters
fig, ax = plt.subplots(figsize=(12,3))
ax.set(title='Daily Precipitation and Snowfall', ylabel='Inches')

# Plot individual lines
precip.plot(style='-')
snow_depth.plot(style='-')

### Create tick locators and formatters
## Minor x-axis ticks
# Locator for every 3 months
loc_3month = mdates.MonthLocator(interval=3)
fmt_months = mdates.DateFormatter("")
## Major x-axis ticks
# Locator for every year
loc_year = mdates.YearLocator(5)
fmt_year = mdates.DateFormatter("%Y")

## Apply tick formatters
# Minor ticks
ax.xaxis.set_minor_locator(loc_3month)
ax.xaxis.set_minor_formatter(fmt_months)
# Major ticks
ax.xaxis.set_major_locator(loc_year)
ax.xaxis.set_major_formatter(fmt_year)

# Finish figure
ax.set_xticks(ax.get_xticks())
fig.autofmt_xdate(which='major', rotation=45, ha='center')
ax.legend(bbox_to_anchor=[1,1]);
```
