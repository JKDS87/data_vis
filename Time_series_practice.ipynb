**Load data and imports**

```python
import pandas as pd
url="https://docs.google.com/spreadsheets/d/e/2PACX-1vT_jChgNsQbHbg4TGepzIqk8XC9DTIKmyyxb1upo5cfZCgbfIUQc2ZC0YMzuU5uApP140Ob49KBjdqh/pub?gid=1198589591&single=true&output=csv"
df = pd.read_csv(url)
df.info()
df.head(3)
```

**1. Convert date column to datetime type**

```python
df['datetime'] = pd.to_datetime(df['date'], format='%Y%m%d')
```

```python
df.info()
df.head(3)
```

**2. Set datetime as new index and inspect**

```python
df = df.set_index(['datetime'])
df.head()
```

**3. Filter down to the requested data**
- preceipitation
- mean_temp
- min_temp
- max_temp
- snow_depth

```python
df = df.drop(columns=['date', 'cloud_cover', 'sunshine', 'global_radiation', 'pressure'])
df = df.loc['2000':]
df.head()
```

**4. Answer the following using .idxmax() and .idxmin(), print, and strftime() methods.**
- A. 'The lowest temperature of {min_temp} degrees occured on {date as MonthName 2-digit day, 4-digit year} which was a {day of the week}.'
- B. 'The highest daily precipitation of {max_precipitation} inches occured on {date as MonthName 2-digit day, 4-digit year} which was a {day of the week}.'

```python
date_min_temp = df['min_temp'].idxmin()
min_temp = df['min_temp'].min()

print(f'The lowest temperature of {min_temp} degrees occured on {date_min_temp.strftime("%B %d, %Y")} which was a {date_min_temp.day_name()}.')
```

```python
date_max_precip = df['precipitation'].idxmax()
max_precip = df['precipitation'].max()

print(f'The highest daily precipitation of {max_precip} inches occured on {date_max_precip.strftime("%B %d, %Y")} which was a {date_max_precip.day_name()}.')
```
